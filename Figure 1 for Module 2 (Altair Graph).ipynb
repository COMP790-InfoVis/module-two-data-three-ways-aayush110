```python
import pandas as pd

data = pd.read_csv("los angeles 2023-01-01 to 2023-03-01 copy.csv")
data.head()
```

```python
import altair as alt

interval = alt.selection_interval(encodings=['y'])

alt.Chart(data).mark_point().encode(
    x='temp:Q',
    y='windspeed:Q',
    color=alt.condition(interval, 'datetime', alt.value('lightgray'))
).add_selection(
    interval
)
```

```python
click = alt.selection_multi(encodings=['color'])

scatter = alt.Chart(data).mark_point().encode(
    x='temp:Q',
    y='windspeed:Q',
    color='datetime:N'
).transform_filter(
    click
)

hist = alt.Chart(data).mark_bar().encode(
    x='temp',
    y='datetime',
    color=alt.condition(click, 'datetime', alt.value('lightgray'))
).add_selection(
    click
)

scatter & hist
```
